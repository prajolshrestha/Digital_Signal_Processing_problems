```python
import numpy as np
import matplotlib.pyplot as plt
import pylab as pl
from IPython import display
import time
```

## sine wave

```python
# 2 Hz sine wave with different sampling rate

sampling_rate = [1,25,50,1000] #Hz
freq = 2
for i,v in enumerate(sampling_rate):
  time = np.arange(0,1,1/v)
  sine = np.sin(2*np.pi*freq*time)

  #plot
  loc = np.unravel_index(i,[2,2],'F')
  plt.subplot2grid((2,2),(loc[1],loc[0]))

  plt.plot(time,sine)

plt.show()
```
