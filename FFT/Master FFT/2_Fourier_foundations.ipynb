## Complex Numbers

```python
#methods to write complex number
#method 1
z = 4 + 1j*3

#method 2
z2 = complex(4,3)

#method 3
z3 = complex('4+3j')

print(z,z2,z3)
```

```python
#plot
plt.plot(np.real(z), np.imag(z),'ro') # plot point
plt.plot([0,4],[0,3]) # plot line

plt.axis('square')
plt.axis([-5,5,-5,5])
plt.grid(True)
plt.xlabel('Real axis'), plt.ylabel('Imaginary axis')
plt.show()
```

```python
#compute the magnitude of complex number (pythagorean Theorem)

#method 1
mag = np.sqrt(np.real(z)**2 + np.imag(z)**2)

#method 2
mag2 = np.abs(z)

print(mag,mag2)
```

```python
# Compute angle of complex number(Pythagorean Theorem)
#method 1
theta = math.atan(np.imag(z) / np.real(z))
#method 2
theta2 = np.angle(z)

print(theta, theta2)
```

```python
# Euler formula with arbitrary vector magnitude

m = 5 #amplitude or magnitude
k = np.pi/4 #angle = 45
compnum = m*np.exp(1j*k)#generate complex number in euler notation
print(compnum)

#extract magnitude and angle
mag = np.abs(compnum)
phase = np.angle(compnum)

plt.polar([phase,phase],[0,mag])
plt.show()
```

## Sine Waves

```python
# simulation parameters
sampling_rate = 500 #Hz
time = np.arange(0.,2.,1./sampling_rate) #seconds

# sine wave parameters
freq = 3
ampl = 2
phase = np.pi/3

#generate sine wave
sine_wave = ampl * np.sin(2*np.pi*freq*time + phase)

plt.plot(time,sine_wave)
plt.grid(True)
plt.show()
```

```python
# sine and cosine are same but phase shift

#generate sine wave

sine_wave = ampl * np.sin(2*np.pi*freq*time + phase)
cosine_wave = ampl * np.cos(2*np.pi*freq*time + phase)

plt.plot(time,sine_wave,'k',label='Sine')
plt.plot(time,cosine_wave,'r',label='Cosine')
plt.legend()
plt.show()

```

## Dot product

```python
# Two vectors
v1 = [1,2]
v2 = [3,2]

# Compute dot product
dp = np.dot(v1,v2)
print(dp)

plt.plot([0,1],[0,2],'r')
plt.plot([0,3],[0,2],'k')
plt.show()
```

```python
# dot product of sine waves

sampling_rate = 500
time = np.arange(0.,2.,1./sampling_rate)

freq1 = 5
freq2 = 5
#freq2 = 6 #dot product of diff ffreq(in steps of 1 and 0.5) is zero
#freq2 = 4.35 # different freq that are other than in steps of 1 and 0.5 results nonzero dotproduct (because of sampling rate)

ampl1 = 2
ampl2 = 2

phase1 = np.pi/2
phase2 = np.pi/2
#phase2 = 0 #dot product = 0 #ie two sine signals are orthogonal to eachother

sine_wave1 = ampl1 * np.sin(2*np.pi*freq1*time + phase1)
sine_wave2 = ampl2 * np.sin(2*np.pi*freq2*time + phase2)

dp = np.dot(sine_wave1,sine_wave2)
print(dp)

```

```python
# dot product with signal and different sine waves

## Signal
theta = 0*np.pi/4 #dot product totally depends on phase, Huge Problem!
#theta = 2*np.pi/4
sampling_rate = 1000
time = np.arange(-1.,1.,1./sampling_rate)

signal = np.multiply(np.sin(2*np.pi*5*time + theta) , np.exp( (-time**2)/ .1))
#plot signal
plt.plot(time,signal)
plt.xlabel('Time'), plt.ylabel('Amplitude')
plt.title("Signal")
plt.show()

## different Sine waves frequencies
sine_freq = np.arange(2.,10.,.5)

## dot product betn signal and different sine waves
dps = np.zeros(len(sine_freq))
for i,v in enumerate(sine_freq):
  sine_wave = np.sin(2*np.pi*v*time)
  dps[i] = np.round(np.dot(sine_wave, signal) / len(time) , 3) #round  for computer errors

plt.stem(sine_freq,dps)
plt.xlabel('sine wave frequency')
plt.ylabel('Dot product')
plt.show()
```

## Complex Dot Product

```python
# phase of signal
theta = 2*np.pi/4; #now for complex sine waves phase does not matter!


# simulation parameters
srate = 1000;
time  = np.arange(-1.,1.,1./srate)

# signal
sinew  = np.sin(2*np.pi*5*time + theta)
gauss  = np.exp( (-time**2) / .1);
signal = np.multiply(sinew,gauss)


# sine wave frequencies
sinefrex = np.arange(2.,10.,.5);

# plot signal
plt.plot(time,signal)
plt.xlabel('Time (sec.)'), plt.ylabel('Amplitude (a.u.)')
plt.title('Signal')
plt.show()



# initialize dot products vector
dps = np.zeros(len(sinefrex),dtype=complex)

# loop over sine waves
for fi in range(len(dps)):

    # create complex sine wave
    sinew = np.exp( 1j*2*np.pi*sinefrex[fi]*time )

    # compute dot product
    dps[fi] = np.abs( np.vdot( sinew,signal )  / len(time))


# and plot
plt.stem(sinefrex,dps)
plt.xlabel('Sine wave frequency (Hz)'), plt.ylabel('Dot product')
plt.title('Dot products with sine waves')
plt.show()
```

### Complex Sine wave Vs. Sine Wave
    (complex dot product Vs. Dot product)

```python
## illustration of the effect of phase offsets on dot products

# create complex sine waves
csw = np.exp(1j*2*np.pi*5*time)
#create real sine waves
rsw = np.sin(   2*np.pi*5*time)

# range of phase offset for signal
phases = np.linspace(0,7*np.pi/2,num=100)

for phi in range(0,len(phases)):

    #create signal
    sinew = np.sin(2*np.pi*5*time + phases[phi])#different phase
    gauss = np.exp((-time**2)/.1)
    signal = np.multiply(sinew,gauss)

    # compute complex dot product
    cdp = np.sum(np.multiply(signal,csw))/len(time)
    #compute real valued dot product
    rdp = np.sum(np.multiply(signal,rsw)) / len(time)

    #plot
    pl.cla()
    plt.subplot2grid((2,2),(0,0),colspan=2)
    plt.plot(time,signal)
    plt.plot(time,rsw)
    plt.title('Signal and sine wave over time')

    # plot complex dot product
    plt.subplot2grid((2,2), (1, 0))
    plt.plot(np.real(cdp),np.imag(cdp),'ro')
    plt.xlabel('Real')
    plt.ylabel('Imaginary')
    plt.axis('square')
    plt.grid(True)
    plt.axis([-.2,.2,-.2,.2])
    plt.plot([0,np.real(cdp)],[0,np.imag(cdp)],'r')


    # draw normal dot product
    plt.subplot2grid((2,2), (1, 1))
    plt.plot(rdp,0,'ro')
    plt.xlabel('Real')
    plt.axis('square')
    plt.grid(True)
    plt.axis([-.2,.2,-.2,.2])

    #show plot
    display.clear_output(wait=True)
    display.display(pl.gcf())
    ttime.sleep(.01)

```
